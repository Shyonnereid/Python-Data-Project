```python
df = pd.read_csv('Bestsellers with categories.csv')
```

```python
df.sort_values('Reviews', inplace=True)
```

```python
df.drop_duplicates(subset= 'Reviews', keep=False, inplace=True)
```

```python
df
```

```python
top_10_author=df.groupby('Author')['Author'].agg(['count']).reset_index().sort_values(by='count',ascending=False).head(10)
print(top_10_author)
```

```python
plt.figure(figsize=(10,4))
df1=sns.barplot(data=top_10_author,x="Author",y="count")
df1.set_title('Name of authors who wrote the most amount books')
plt.xticks(rotation=90)
plt.show()
```

```python
Top_10_bestselling_book=df.groupby('Name')['Name'].agg(['count']).reset_index().sort_values(by='count',ascending=False).head(10)
print(Top_10_bestselling_book)
```

```python
plt.figure(figsize=(10,4))
df1=sns.barplot(data=Top_10_bestselling_book,x="Name",y="count")
df1.set_title('Top 10 Best-selling books')
plt.xticks(rotation=90)
plt.show()
```
